```python
from keras.models import Model
```

```python
from keras.layers import *

x = Input(shape=(3,))

hid1 = Dense(128, activation='relu')(x)
hid2 = Dense(64, activation='relu')(hid1)

out1 = Dense(1, activation='linear')(hid2)
out2 = Dense(1, activation='sigmoid')(hid2)
```

```python
model = Model(inputs = x, outputs = [out1, out2])
```
